```python
x, y = sm.symbols("x, y")
```

```python
b_vec = sm.Matrix([3, 0])
b_vec
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


def random_vec2d(n):
    x = np.linspace(-10, 10, n) + np.random.uniform(-10, 10, size=n)
    y = np.linspace(-10, 10, n) + np.random.uniform(-10, 10, size=n)  # noqa: NPY002
    vecs = [np.array([i, j]) for i, j in zip(x, y)]  # noqa: B905
    return vecs


def plot_vec2d(vector2d, ax, color=np.random.random(3), **options):
    return ax.arrow(
        0,
        0,
        vector2d[0],
        vector2d[1],
        head_width=0.3,
        head_length=0.3,
        length_includes_head=True,
        color=color,
        **options,
    )
```

```python
n = 1_000
vecs = random_vec2d(n)

fig, ax = plt.subplots(figsize=(13, 8))
for vec in vecs:
    plot_vec2d(vec, ax=ax, color=np.random.random(3), linewidth=4)

plt.axis([-20, 20, -20, 20])
plt.axis("off")
plt.title(r"Random Vectors in $\mathbb{R}^2$", fontdict={"size": 20})
```

```python
def l2_norm(vector):
    return (np.sum([i**2 for i in vector])) ** (1 / 2)


l2_normed_vecs = [vec / l2_norm(vec) for vec in vecs]

fig, ax = plt.subplots(figsize=(9, 6))
for vec in l2_normed_vecs:
    plot_vec2d(vec, ax=ax, color=np.random.random(3), linewidth=0.1)

plt.axis([-2, 2, -2, 2])
plt.title("L2 Normalized Vectors", fontdict={"size": 20})
plt.xlabel("$x$", fontdict={"size": 25})
plt.ylabel("$y$", fontdict={"size": 25})
```

```python
def l1_norm(vec):
    return np.sum([np.abs(i) for i in vec])


l1_normed_vecs = [vec / l1_norm(vec) for vec in vecs]

fig, ax = plt.subplots(figsize=(9, 6))
for vec in l1_normed_vecs:
    plot_vec2d(vec, ax=ax, color=np.random.random(3), linewidth=0.1)

plt.axis([-2, 2, -2, 2])
plt.title("L1 Normalized Vectors", fontdict={"size": 20})
plt.xlabel("$x$", fontdict={"size": 25})
plt.ylabel("$y$", fontdict={"size": 25})
```

```python
def inf_norm(vec):
    return np.max([abs(i) for i in vec])


inf_norm_vecs = [vec / inf_norm(vec) for vec in vecs]

fig, ax = plt.subplots(figsize=(9, 6))
for vec in inf_norm_vecs:
    plot_vec2d(vec, ax=ax, color=np.random.random(3), linewidth=0.1)

plt.axis([-2, 2, -2, 2])
plt.title(r"$\infty$ Normalized Vectors", fontdict={"size": 20})
plt.xlabel("$x$", fontdict={"size": 25})
plt.ylabel("$y$", fontdict={"size": 25})
```
