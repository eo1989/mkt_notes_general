```python
def describe_df(df: pd.DataFrame) -> pd.DataFrame:
    print(f"The dataset contains {df.shape[1]} columns & {len(df)} rows.")

    for col in df.columns:
        col_dtype = df[col].dtype

        print(f"\nColumn: {col} ({col_dtype})")
        if col_dtype == "object":
            print(
                f"--- Percentage of NaNs: {df[col].isna().sum() / len(df[col]) * 100}"
            )
            print(f"--- Unique values: \n {df[col].unique()}")
        else:
            print(f"--- Summary statistics:\n {df[col].describe()}")
```
