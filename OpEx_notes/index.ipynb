```python
def filter_opts_data(options, by: Literal["expiration", "strike"] = "expiration"):
    data = pd.DataFrame()
    data["Total OI"] = options.groupby(by)["open_interest"].sum()
    data["Call OI"] = (
        options[options["option_type"] == "call"].groupby(by)["open_interest"].sum()
    )
    data["Put OI"] = (
        options[options["option_type"] == "put"].groupby(by)["open_interest"].sum()
    )
    data["Total Volume"] = options.groupby(by)["volume"].sum()
    data["Call Volume"] = (
        options[options["option_type"] == "call"].groupby(by)["volume"].sum()
    )
    data["Put volume"] = (
        options[options["option_type"] == "put"].groupby(by)["volume"].sum()
    )
    return data


data = filter_opts_data(options, "strike")
data
```
